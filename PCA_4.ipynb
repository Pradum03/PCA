```python
from sklearn.preprocessing import LabelEncoder
type_encoded = LabelEncoder()
df['type'] = type_encoded.fit_transform(df['type'])
```

```python
df['type'].unique()
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
df_std= std.fit_transform(df)
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(20, 6))

sns.boxplot(df_std)
```

```python
df_std = pd.DataFrame(data=df_std, columns=df.columns.tolist())
```

```python
df_std.head()
```

```python
df['quality'].unique()
```

```python
X = df_std.iloc[:,:-1]
y = df_std.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca.fit(X)
pca_data = pca.transform(X)
df_pca = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2','PC3'])
df_pca['target_variable'] = y.values
print(df_pca.head())


```

```python
explained_variance_ratio = pca.explained_variance_ratio_

# Plot the explained variance ratio for each component
plt.plot(range(1, len(explained_variance_ratio)+1), explained_variance_ratio)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_data[:,0], pca_data[:,1], pca_data[:,2])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
# Perform K-Means clustering on the PCA-transformed data with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(pca_data)
labels = kmeans.labels_

# Create a scatter plot of the first three principal components, colored by cluster
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(pca_data[:,0], pca_data[:,1], pca_data[:,2], c=labels)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Split the PCA-transformed data and cluster labels into training and test sets
X_train, X_test, y_train, y_test = train_test_split(pca_data, labels, test_size=0.3, random_state=42)

# Train a logistic regression model on the training set
logreg = LogisticRegression(random_state=42)
logreg.fit(X_train, y_train)

# Make predictions on the test set using the trained model
y_pred = logreg.predict(X_test)

# Evaluate the accuracy of the model on the test set
accuracy = logreg.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```
